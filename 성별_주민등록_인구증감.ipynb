```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data = pd.read_csv('./data/행정안전부_지역별(행정동)_성별_주민등록_인구증감_20240930.csv', encoding='cp949')
data
```

```python
data['행정기관코드'].dtype
```

```python
data['행정기관코드'] = data['행정기관코드'].astype(str)
```

```python
data['행정기관코드'].dtype
```

```python
# '행정기관코드'의 앞 2자리가 '41'인 것만 남기고 data에 덮어쓰기
data = data[data['행정기관코드'].str.startswith('41')]

# 결과 확인
data
```

```python
data['기준연월'].dtype
```

```python
data['기준연월'] = pd.to_datetime(data['기준연월'], format='%Y-%m-%d').dt.to_period('M')
```

```python
data
```

```python
data.columns
```

```python
data = data.drop(['시도명',  '전체 전월인구수', '전월 남자인구수',
       '전월 여자인구수', '전체 당월인구수', '당월 남자인구수', '당월 여자인구수'], axis=1)
```

```python
data
```

```python
data['통합주소'] = data['시군구명'] + " " + data['읍면동명']
data
```

```python
data.columns
```

```python
data = data[['행정기관코드', '기준연월', '시군구명', '읍면동명', '통합주소', '전체 인구증감', '남자 인구증감', '여자 인구증감']]
data
```

```python
data['행정기관코드'] = data['행정기관코드'].astype(np.int64)
```

```python
data['행정기관코드'].dtype
```

```python
df = pd.DataFrame(data)
```

```python
# 저장할 경로 수정
output_dir = "C:/fintech_service/Project/data/전처리"
output_file = "성별_주민등록_인구증감_전처리.csv"

# 폴더가 존재하지 않으면 생성
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 데이터프레임을 CSV 파일로 저장
df.to_csv(os.path.join(output_dir, output_file), index=False, encoding="utf-8-sig")
```
