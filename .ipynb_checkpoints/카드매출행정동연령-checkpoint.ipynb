```python
data['etl_ym'] = pd.to_datetime(data['etl_ym'], format='%Y%m')
```

```python
data['etl_ym'] = pd.to_datetime(data['etl_ym'], format='%Y%m').dt.to_period('M')
```

```python
# admi_nm과 cty_nm이 문자열 형식인지 확인
data['cty_nm'] = data['cty_nm'].astype(str)
data['admi_nm'] = data['admi_nm'].astype(str)

# 두 컬럼을 결합하여 city 컬럼 생성
data['city'] = data['cty_nm'].str.strip() + " " + data['admi_nm'].str.strip()
```

```python
data = data[['etl_ym', 'admi_cd', 'admi_nm', 'cty_cd', 'cty_nm', 'city', 'buz_type_cd',
       'buz_type_nm', 'card_age_m10_19_amt', 'card_age_m20_29_amt',
       'card_age_m30_39_amt', 'card_age_m40_49_amt', 'card_age_m50_59_amt',
       'card_age_m60_up_amt', 'card_age_f10_19_amt', 'card_age_f20_29_amt',
       'card_age_f30_39_amt', 'card_age_f40_49_amt', 'card_age_f50_59_amt',
       'card_age_f60_up_amt', 'card_age_tot_amt', 'card_age_m_tot_amt',
       'card_age_f_tot_amt']]
```
