```python
def preprocess_data(data):
    # 불필요한 열 삭제
    data = data.drop(['hour', 'day', 'cnt'], axis=1)
    
    # 날짜 형식 변환 및 연월 컬럼 생성
    data['ta_ymd'] = pd.to_datetime(data['ta_ymd'], format='%Y%m%d')
    data['year_month'] = data['ta_ymd'].dt.to_period('M').dt.strftime('%Y-%m')  # YYYY-MM 형식으로 변환
    
    # 원래 날짜 컬럼 삭제
    data = data.drop(['ta_ymd'], axis=1)
    
    # 열 순서 재정렬
    data = data[['year_month', 'cty_rgn_no', 'admi_cty_no', 'card_tpbuz_cd', 
                 'card_tpbuz_nm_1', 'card_tpbuz_nm_2', 'sex', 'age', 'amt']]
    
    # 'card_tpbuz_nm_1'에서 '음식'만 남기고 필터링
    data = data[data['card_tpbuz_nm_1'] == '음식']
    
    # 필요 없는 컬럼 최종 삭제
    data = data.drop(['card_tpbuz_nm_1'], axis=1)
    
    return data
```

```python
# 모든 ZIP 파일 경로 가져오기
zip_files = glob.glob('./data/카드소비데이터/*.zip')

# 데이터를 담을 빈 리스트 생성
data_frames = []

# 각 ZIP 파일을 읽어와 전처리 후 데이터프레임 리스트에 추가
for zip_file in zip_files:
    with zipfile.ZipFile(zip_file, 'r') as z:
        for file in z.namelist():
            if file.endswith('.csv'):  # CSV 파일만 처리
                with z.open(file) as f:
                    df = pd.read_csv(f)
                    processed_df = preprocess_data(df)
                    data_frames.append(processed_df)

# 모든 전처리된 데이터프레임을 하나로 병합
final_data = pd.concat(data_frames, ignore_index=True)

# 결과 확인
print(final_data.head())
```

```python
# 전처리된 데이터를 CSV 파일로 저장
output_path = './data/processed_data.csv'
final_data.to_csv(output_path, index=False)

print(f"데이터가 '{output_path}'에 저장되었습니다.")
```

# 연도별로 잘라서 저장

```python
# 모든 ZIP 파일 경로 가져오기
zip_files = glob.glob('./data/카드소비데이터/*.zip')

# 출력 디렉토리 설정
output_dir = './data/processed_chunks'
os.makedirs(output_dir, exist_ok=True)

# 각 ZIP 파일을 읽어와 전처리 후 저장
for zip_file in zip_files:
    with zipfile.ZipFile(zip_file, 'r') as z:
        for file in z.namelist():
            if file.endswith('.csv'):  # CSV 파일만 처리
                with z.open(file) as f:
                    df = pd.read_csv(f)
                    processed_df = preprocess_data(df)
                    
                    # 연도별로 그룹화하여 분기별로 나눠 저장
                    for year, year_data in processed_df.groupby(processed_df['year_month'].str[:4]):
                        for quarter in range(1, 5):
                            # 각 분기 설정 (1분기: 1~3월, 2분기: 4~6월, 3분기: 7~9월, 4분기: 10~12월)
                            if quarter == 1:
                                quarter_data = year_data[year_data['year_month'].str[-2:].isin(['01', '02', '03'])]
                            elif quarter == 2:
                                quarter_data = year_data[year_data['year_month'].str[-2:].isin(['04', '05', '06'])]
                            elif quarter == 3:
                                quarter_data = year_data[year_data['year_month'].str[-2:].isin(['07', '08', '09'])]
                            elif quarter == 4:
                                quarter_data = year_data[year_data['year_month'].str[-2:].isin(['10', '11', '12'])]

                            # 비어있지 않은 경우만 저장
                            if not quarter_data.empty:
                                output_path = f'{output_dir}/processed_data_{year}_Q{quarter}.csv'
                                quarter_data.to_csv(output_path, mode='a', index=False, header=not os.path.exists(output_path))
                                print(f"{year}년 {quarter}분기 데이터가 {output_path}에 저장되었습니다.")
```
